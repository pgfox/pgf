# Plot Testing Notebook

Use this notebook to experiment with the plotting helpers in `pgf.plot`.

```python
import pandas as pd
import numpy as np

from pgf.plot import qq_plot, histogram_plot, histogram_bin_counts, cdf_plot

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sample = pd.Series(np.random.normal(loc=0, scale=1, size=10000))

hist_counts = histogram_bin_counts(sample)
hist_counts
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
histogram_plot(sample, bins=hist_counts['sturges'], ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
histogram_plot(sample, bins=hist_counts['sturges'], ax=ax, density=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
qq_plot(sample, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
cdf_plot(sample, ax=ax)
plt.show()
```

```python
normal_sample = pd.Series(np.random.normal(loc=0, scale=1, size=500))
uniform_sample = pd.Series(np.random.uniform(low=-3, high=3, size=500))

fig, ax = plt.subplots(figsize=(6, 4))
cdf_plot(normal_sample, ax=ax)
cdf_plot(uniform_sample, ax=ax)
ax.legend(['Normal', 'Uniform'])
ax.set_title('CDF: Normal vs Uniform')
plt.show()
```
